```python
import pandas as pd

df= pd.read_csv("Biodata3.csv")
print(df)
```

```python
df.dropna(axis=1) #missing values column's remove
```

```python
df.dropna(axis=0) #missing values rows's remove
```

```python
df.dropna(how="any") #remove missing values's cols
```

```python
df.dropna(how="all") #remove missing values's rows
```

```python
df.dropna(subset=["Gender"])
```

```python
df.fillna("PYTHON") #filling null values with PYTHON using fillna method
```

```python
df.fillna({"Name":404,"Gender":"FMFM","Mobile":"NaN","Id":"NaN","Address":404,"Nationality":404}) #You can replace perticular column values
```

```python
df.ffill() #fill missing values from begin value
```

```python
df.bfill() # fill missing values from backword value
```

```python
df.fillna(12,inplace=True) #it is replace value
df
```

```python
df.replace(to_replace=12,value="SURAJ") #REPLACE to all 12 values from SURAJ
```

```python
df.replace(["male"],"SURAJ")
```
